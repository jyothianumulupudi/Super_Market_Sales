```python
supermarket_data = pd.read_excel("/content/Python_mandatory_project_supermarket_sales.xlsx") # reading the data of a super market sales file
supermarket_data.head() #displaying first 5 lines of the dataset
```

#1.Is your supermarket more popular with males or females?



Here we have to calculate supermarket is more popular with males or females. Here column 'Gender' is a categorical data.So we can use count plot for this analysis.

```python
sns.countplot(x='Gender', data=supermarket_data) # x is an column 
plt.title("Figure1: Count of Male and Female") # title for the plot
plt.xlabel("Gender") # x-axis label
plt.ylabel("Count") # y-axis label 
plt.show()
print(supermarket_data['Gender'].value_counts()) #count of the male and female.
```

Above graph explains population of the male and female.Countplot is basically a count of frequency plot in the form of a bar graph. It plots the count of each category in a separate bar.People from both the genders are making almost same number of purchases. But the female(501) is a bit hight than the male(409).

#2.What does the customer rating look like and can you also comment on its distribution? 


```python
sns.distplot(a=supermarket_data['Rating'], color='m',bins=10, hist_kws={"edgecolor": 'white'})# distplot for customer rating
plt.title("Figure2: Distribution of Customer Rating") # title of the plot
plt.xlabel("Rating") # x-aixs label
plt.ylabel("Density") # y-axis label
plt.show()
```

The above figure explains us about the distribution of the customer rating.The data is normally distributed. The rating distribution looks uniform and there seems to be no skewness on the left or right side of the distribution

#3.What can you say about the aggregate sales across the branches? 


```python
sns.countplot(x='Branch', data=supermarket_data) # counplot to know the hight sales in a branch
plt.title("Figure3: Aggregate sales of branches") # title for the plot
plt.xlabel("Branches") # x-label title
plt.ylabel("Aggregrate sales") # y-label title
plt.show()
print(supermarket_data['Branch'].value_counts()) # count of the each branch
```

The above figure explains which branch have more sales.Here A(340) is little grater than the remaining braches B(332), C(328).

#4.Which is the most popular payment method used by customers?

```python
plt.figure(figsize=(6,6))
sns.countplot(x='Payment', data=supermarket_data) #countplot to know the most popular payment method.
plt.title("Figure4: Popular Payment Menthod") # title of the graph
plt.xlabel("Payment methods") # x-axis title
plt.ylabel("Total Count") # y- axis title
plt.grid(True) # we can show the grids for clear understanding
plt.show()
print(supermarket_data['Payment'].value_counts())
```

The above figure tells us about the most popular payment menthod used in the super market is Ewallet. Here Ewallet(345) value is a bit higher than the cash(344) and credit card(311) payments.

#5.Does gross income affect customer ratings?

Here  we have to check the relationship between gross income and customer ratings. Here both are numberic data.Here we can use Bi-variate analysis(interaction/relationship b/w them).So I'm using scatterplot.

```python
sns.regplot(x= "Rating", y= "gross income", data = supermarket_data) # use the function regplot to make a scatterplot for rating and gross income
plt.title("Figure5: Gross Income vs Rating") # title of the plot
plt.xlabel("Ratings of the customer") # x-axis title
plt.ylabel("Gross Income") # y-axis title
plt.show()
```

In the above we can clearly observe trend line is quite flat so it clearly says that gross income does not effect the customer ratings.

#6.Which is the most profitable branch? 


Here we have to find out which is the most profitable branch.Here we have to about the gross income of a every branch, gross income is a numerical data and branch is a categorical data. So I'm going to use box plot.

```python
sns.boxplot(x='Branch', y='gross income', data=supermarket_data)# we can get more profitable branch here
plt.title("Figure6: Mosi Profitable Branch") # title of the plot
plt.xlabel("Branch") # x-axis title
plt.ylabel("Gross Income") # y-axis title
plt.show()
```

The above figure explains us about the most profitable branch.We can clearly observe that both A and B branches have same profit.But C has more profitable branch than A and B.

#7.Create a boxplot between Gender and Gross income and explain the results?

Here we have to calculate which gender brings more profit to the super market.

```python
sns.boxplot(x='Gender', y='gross income', data=supermarket_data) #plotting the data
plt.title("Figure7: Gross Income by Gender") # title of the plot
plt.xlabel("Gender") # x-axis title
plt.ylabel("Gross Income") # y-axis title
plt.show()
```

#8.Can you see any time trend in gross income? 


```python
plt.figure(figsize=(10,8)) # Figure size
sns.lineplot( x = "Date", y = "gross income", data = supermarket_data) # Fucntion to find out trend in a gross income
plt.title("Figure8: Trend in Gross Income") # title of the plot
plt.xlabel("Date") # title of the x-axis
plt.ylabel("Gross Income") # title of the y-axis
plt.show()
```

#9.What is the spending pattern of both males and females, meaning in which category do they spend more?

```python
sns.barplot(x='Product code', y='gross income', data=supermarket_data)
plt.title("Figure9: Gross Income by Products") # title of the plot
plt.xlabel("Products") # title of the x-axis
plt.ylabel("Gross Income") # title of the y-axis
# legend for giving product names to product codes
plt.legend(['PD1: Healthy and beauty', 'PD2: Electronic accessories', 'PD3: Home and lifestyle','PD4: Sports and travel','PD5: Food and beverages', 'PD6: Fashion accessories'], loc ="lower right")
plt.grid(True)
plt.show()
```

#11.Which day of the week has maximum sales? 



```python
plt.figure(figsize=(8, 6))
supermarket_data['weekday'] = (supermarket_data['Date']).dt.day_name() # to get the day of the week
plt.title('Daily Sales by Day of the Week') # title of the week
sns.countplot(supermarket_data['weekday']) # to get the maximum sales of the day
plt.show()
```

The above figure explains us about the which day has more sales.Sales is highest on Saturdays probably because it is the weekend.

#12.Which hour of the day is busiest?



```python
supermarket_data['Hour'] = pd.to_datetime(supermarket_data['Time'], format='%H:%M:%S').dt.hour # to get the hours from the time
sns.lineplot(x="Hour",  y = 'Quantity',data =supermarket_data).set_title("Product Sales per Hour") # line plot to know hour of the day is busiest
plt.show()
```

The value is hight at 14th hour i.e 2 pm of the day. Hence, sales is typically higher in the afternoons.
